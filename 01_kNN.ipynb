**IMPORTANDO BIBLIOTECAS**

```python
#Bibliotecas de manipualção e visualização de dados
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
import seaborn as sns

#Classes do modelo de aprendizado
from sklearn.neighbors import KNeighborsClassifier

#Funções de avaliação dos modelos
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

**ANÁLISE DO CONJUNTO DE DADOS**

```python
#Carregando o dataset
dataset = pd.read_csv('https://raw.githubusercontent.com/JusciAvelino/Monitoria-ModelagemPreditivaDescritiva1//main/Iris.csv')
```

```python
print("Quantas classes existem nesse dataset?\n%d" %(len(dataset["Species"].unique())))
print("\nQuantas instâncias existem no dataset?\n%d" %(len(dataset)))
print("\nQuantas features existem no dataset?\n%d" %(dataset.shape[1]))
print("\nQue features são essas?\n%s" %(str([k for k in dataset.keys()])))
print("\nQual o numero de instâncias por classe?")
print(dataset["Species"].value_counts())
```

```python
# Distribuição das classes pelas features SepalWidthCm e PetalLengthCm

sns.set_style("whitegrid")
sns.FacetGrid(dataset, hue ="Species",
              height = 6).map(plt.scatter,
                              'SepalWidthCm',
                              'SepalLengthCm').add_legend()
```

```python
#Mapeando os valores da classe para inteiro (para fins de visualização da região de decisão)
dataset['Species'] = pd.factorize(dataset['Species'])[0]
mappingSize = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
dataset
```

```python
#Vamos usar somente duas features SepalLengthCm e SepalWidthCm
X = dataset.iloc[:,[0,1]]
y = dataset.iloc[:,[4]]
```

**TREINAMENTO DO MODELO**

```python
#Separando o conjunto de dados em treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y)

model = KNeighborsClassifier()
#treinando o modelo
model.fit(X_train, y_train)

#predição
y_pred = model.predict(X_test)

#Resultados do classificador
print(classification_report(y_test, y_pred))
```

**REGIÃO DE DECISÃO**

```python
def show_decision_region(x, y, clf):
    plot_decision_regions(x, y, clf=clf)
    plt.xlabel("SepalLengthCm")
    plt.ylabel("SepalWidthCm")
    if clf.__class__.__name__ == "KNeighborsClassifier":
        plt.title(clf.__class__.__name__ + " k = " + str(clf.n_neighbors))
    else:
        plt.title(clf.__class__.__name__)
    plt.show()

show_decision_region(np.array([X_test["SepalLengthCm"].values, X_test["SepalWidthCm"].values]).T, y_test.values.reshape(-1,1).T[0], model)
```

**VARIANDO O TAMANHO DA VIZINHANÇA**

```python
for k in [11, 9, 7, 5, 3, 1]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)

    show_decision_region(
        np.stack(
            [
                X_test["SepalLengthCm"].values, 
                X_test["SepalWidthCm"].values,
            ],
            axis=1
        ), 
        y_test.values.reshape(-1,1).T[0],
        model
    )
```

**MEDIDA DE DISTÂNCIA**

```python
for m in ["euclidean", "manhattan"]:
    model = KNeighborsClassifier(metric=m)
    model.fit(X_train, y_train)

    show_decision_region(
        np.stack(
            [
                X_test["SepalLengthCm"].values, 
                X_test["SepalWidthCm"].values,
            ],
            axis=1
        ), 
        y_test.values.reshape(-1,1).T[0],
        model
    )
```
